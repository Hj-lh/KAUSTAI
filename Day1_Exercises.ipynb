## Exercise #1

Expectation: A working solution where the learned functions (for each training dataset) using Linear Regression is close to the training data points.

```python
#[Fictional Story] The astronauts on moon wanted to measure the value of gravity on the moon. The decide to throw a spanner
#from different heights and measure the time needed to cover the distance and based on these values want to extract the value of
#gravity on moon. The values might be slightly off due to human error and the (very thin) moon atmosphere. Use linear regression to
#find the value of gravity on the moon.



import matplotlib.pyplot as plt
import numpy as np

gmoon=9.8/6
t=np.linspace(1,3,20)
s=0.5*gmoon*t**2 + 0.05*np.random.randn(t.size)
plt.plot(t,s,'.')

x= t
y= s
print(x.shape)
print(y.shape)
```

```python
M = 2
X = []
for i in range(M+1):
  X.append(x**i)


X = np.array(X).T

theta = np.linalg.solve(X.T@X, X.T@y)
yhat = X@theta


plt.plot(x,y,'ro', x, yhat)
```

```python
# Assuming you've already calculated yhat using X and theta

# Calculate R-squared
ss_total = np.sum((y - np.mean(y))**2)
ss_residual = np.sum((y - yhat)**2)
r_squared = 1 - (ss_residual / ss_total)

# Convert R-squared to percentage
r_squared_percentage = r_squared * 100

# Print the R-squared as a percentage
print("R-squared (%):", r_squared_percentage)
```

## Exercise #3

Expectation: A working solution where the learned functions (for each training dataset) using Linear Regression is close to the training data points.


```python
#Linear Regression for general functions
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi


x=np.linspace(-1,1,100)

y=np.sin(2*pi*.5*x)+0.3*np.random.randn(x.size)
plt.plot(x,y, 'ro')
```

```python
M = 3

X =[]


x = np.reshape(x, (100,1))
basis = np.arange(M+1)
basis = np.reshape(basis, (1, M+1))
X = x**basis

print(X.shape)

theta = np.linalg.inv(X.T @ X) @ (X.T @y)

yhat = X@ theta

plt.plot(x,y,'bo', x, yhat)
```
